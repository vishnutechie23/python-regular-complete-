```python
import turtle
import time

t = turtle.Turtle()
t.color("red")
t.width(7)

def draw_heart():
    t.begin_fill()
    t.left(140)
    t.forward(180)
    t.circle(-90, 200)
    t.left(120)
    t.circle(-90, 200)
    t.forward(180)
    t.end_fill()

# Animation loop (beating effect)
for i in range(1):
    t.clear()
    t.penup()
    t.goto(0, -50)
    t.pendown()
    t.begin_fill()
    draw_heart()
    t.end_fill()
    t.width(10 + i)  # change size effect
    time.sleep(1)

turtle.done() 
```

```python
import turtle
import random
import time

screen = turtle.Screen()
screen.bgcolor("black")

t = turtle.Turtle()
t.hideturtle()
t.speed(0)
t.color("red")

# ❤️ heart drawing function
def draw_heart(size):
    t.begin_fill()
    t.left(140)
    t.forward(size)
    t.circle(-size/2, 200)
    t.left(120)
    t.circle(-size/2, 200)
    t.forward(size)
    t.end_fill()
    t.setheading(0)

# 💓 BEATING BIG HEART
for scale in range(80, 120, 3):
    t.clear()
    t.penup()
    t.goto(0, -50)
    t.pendown()
    draw_heart(scale)
    time.sleep(0.05)

for scale in range(120, 80, -3):
    t.clear()
    t.penup()
    t.goto(0, -50)
    t.pendown()
    draw_heart(scale)
    time.sleep(0.05)

# 💥 SPLIT INTO SMALL HEARTS
hearts = []

for _ in range(50):
    hearts.append({
        "x": 0,
        "y": 0,
        "dx": random.uniform(-5, 5),
        "dy": random.uniform(-5, 5)
    })

# animation loop
for _ in range(80):
    t.clear()

    for h in hearts:
        h["x"] += h["dx"]
        h["y"] += h["dy"]

        t.penup()
        t.goto(h["x"], h["y"])
        t.pendown()
        draw_heart(10)

    time.sleep(0.05)

turtle.done() 
```

```python
import turtle
import random
import time

screen = turtle.Screen()
screen.bgcolor("black")
screen.setup(width=800, height=600)

t = turtle.Turtle()
t.hideturtle()
t.speed(0)
t.color("red")

# ❤️ Heart function
def draw_heart(size):
    t.begin_fill()
    t.left(140)
    t.forward(size)
    t.circle(-size/2, 200)
    t.left(120)
    t.circle(-size/2, 200)
    t.forward(size)
    t.end_fill()
    t.setheading(0)

# 💓 Big heart beat
for scale in range(80, 120, 3):
    t.clear()
    t.penup()
    t.goto(0, -50)
    t.pendown()
    draw_heart(scale)
    time.sleep(0.03)

for scale in range(120, 80, -3):
    t.clear()
    t.penup()
    t.goto(0, -50)
    t.pendown()
    draw_heart(scale)
    time.sleep(0.03)

# 💥 Create scattered hearts
hearts = []
for _ in range(50):
    hearts.append({
        "x": 0,
        "y": 0,
        "dx": random.uniform(-6, 6),
        "dy": random.uniform(-6, 6)
    })

# 💫 Scatter animation with bounce
while True:
    t.clear()

    for h in hearts:
        # move
        h["x"] += h["dx"]
        h["y"] += h["dy"]

        # bounce from edges
        if h["x"] > 390 or h["x"] < -390:
            h["dx"] *= -1
        if h["y"] > 290 or h["y"] < -290:
            h["dy"] *= -1

        # draw heart
        t.penup()
        t.goto(h["x"], h["y"])
        t.pendown()
        draw_heart(10)

    time.sleep(0.03)

turtle.done()
```
