## Operator : assingment


```python
a = 20
b = 20
print(f"a = {a} , b = {b}")
a += b 
print(f"a = {a} , b = {b}")
a -= b
print(f"a = {a} , b = {b}")
a *=b
print(f"a = {a} , b = {b}")
a /=b
print(f"a = {a} , b = {b}")
a //= b
print(f"a = {a} , b = {b}")
a %= b
print(f"a = {a} , b = {b}")
a **= b
print(f"a = {a} , b = {b}")

```

## Operator : Bitwise


```python
a = 26
b = 20

print(a & b)
print(a | b)
print(a << b)
print(a >> b)
print(~ b)
```

## membership : Operator 


```python
li = [8, 10, 2 ,30, 4]
print(f"List : {li}")
print(f"8 in {li} : {8 in li}")
print(f"10 in {li} : {10 in li}")
print(f"100 in {li} : {100 in li}")
```
