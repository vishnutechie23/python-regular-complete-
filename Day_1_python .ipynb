## Operators : Arithematic Operators

```python
a = 12 
b = 5 

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
```

## Operators : Comparison Operator 

```python
a = 10 
b = 10 

print(f"a = {a}, b = {b}")
print(f"{a} == {b} : {a == b}")
print(f"{a} != {b} : {a != b}")
print(f'{a} < {b} : {a < b}')
print(f'{a} > {b} : {a > b}')
print(f'{a} <= {b} : {a <= b}')
print(f'{a} >= {b} : {a >= b}')
```

## print methods 

```python

p = 100000
r = 10
t = 60 
si = p * r * t / 100 * 12
```

```python
#1. normal method
print("The simple Interest for" ,p,  "at" ,r, "% rate of interest for" ,t, "months is" ,si ) 
```

```python
# f-string mwthod (formatting string literal)
print(f"The simple Interest for {p}  at {r} % rate of interest for {t} moths is {si} ") 
```

```python
#3. Format method 
print( "The simple Interest for {} at {} % rate of interest for {} months is {}" .format (p,r,t,si)) 
```
