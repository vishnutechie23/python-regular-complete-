## Operators : Assignment 

```python
a = 10 
b = 20 

print(f"a = {a}, b = {b}") 
a += b 
print(f"a = {a}, b = {b}") 
a -= b 
print(f"a = {a}, b = {b}") 
a *= b 
print(f"a = {a}, b = {b}") 
a /= b 
print(f"a = {a}, b = {b}") 
a //= b 
print(f"a = {a}, b = {b}") 
a %= b 
print(f"a = {a}, b = {b}") 
b  **= a 
print(f"a = {a}, b = {b}")  
```

## Operators : Bitwise 

```python
a = 10 
b = 25 

print(a & b) 
print(a | b)
print(b << 1) 
print(b >> 2)
print(~a)  

```
