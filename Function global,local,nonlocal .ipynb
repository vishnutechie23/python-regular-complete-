```python
# ncr 

n = int(input()) 
r = int(input()) 

n_fact, r_fact, n_r_fact = 1, 1, 1
# n_fact = r_fact = n_r_fact = 1 

for i in range(1, n+1):
    n_fact *= i 
for i in range(1, r + 1):
    r_fact *= i
for i in range(1, n - r + 1):
    n_r_fact *= i 

ans = n_fact / (r_fact * n_r_fact) 

print(f"{n}C{r} = {ans}")
```

```python
def factorial(n : int) -> int:
    ans = 1 

    for i in range(2, n+1):
        ans *= i 
    
    return ans 
```

```python
n = int(input("n = "))
r = int(input('r = '))

ans = factorial(n) / (factorial(r) * factorial(n-r))

print(f"{n}C{r} = {ans}")
```

```python
# Prime Number 

def isPrime(n : int) -> bool:
    if n < 2:
        return False 
    
    for i in range(2, n):
        if n % i == 0:
            return False 
    else:
        return True 
```

```python
for i in range(1, 101, 1):
    print(f"{i} --> {'Prime' if isPrime(i) == True else 'Not Prime'}")
```

```python
for i in range(1, 51, 1):
    if isPrime(i):
        print(i, end = " ") 
```

```python
def new():
    return a + 2 

a = 10 
print(f"Current Value : a = {a}") 
a = new() 
print(f"Updated Value : a = {a}") 
```

```python
def final(a):
    print(f"Original Variable (Global) : a = {a}")
    # global a 
    a = a + 2 
    print(f"Local Variable : a = {a}") 
a = 10 
print(f"Original Value : a = {a}")
final(a) 
print(f"Updated Value : a = {a}")
```

```python
def f2(a):
    print(f"Global a = {a}, Memory Location : {id(a)}") 
    # local a = global a + 2 
    a = a + 2 
    print(f"Local a = {a}, Memory Location : {id(a)}") 

a = 10 # Global 
print(f"Global a = {a}")  
f2(a) 
print(a) 
```

```python
def f3():
    y = 20 
    def f4():
        global x 
        x += 50         
        print(x)  
    f4() 

x = 100 
print(x) 
f3() 
print(x) 
```

```python
def f5():
    y = 40 
    print(y)  
    def f6():
        nonlocal y 
        y += 20 
        print(y) 
    
    f6() 
    print(y) 

x = 100 
print(x) 
f5() 
```
