## LIsts


```python
#List creation
a = [1,2,3,4]
b = list((1,2,3))
c = []
d = list() 
print(a,type(a))
print(b,type(b))
print(c,type(c))
print(d,type(d))
```

```python
#indexing : forward/negative(backward)
a = [1,2,3,4,5,6]
print(a[0])
print(a[2])
print(a[-3])
print(a[-1])
```

```python
#Slicing : slicer default values (start=0,stop=len,step=1)
li = [10,20,30,40,50,60,70,80]
print(li[0:4:1])
print(li[0::])
print(li[1::2])
print(li[::-1]) #reverse 
```

```python
# UPdating : access -> assign and update
a = [1,2,3,4]
a[0] = 1000
a[1] = 2000
print(a)
```

```python
#DElete elements 
# Remove() : remove by valuee
a = [1,2,3,4,5,6,7,8] 
print(a)
a.remove(2)
print(a)

#pop() : Remove by index 
a.pop() # last 
a.pop(0) # First 
print(a)

#del 
del a[1]
print(a)
```

```python
# append() & extend() 

# append : add whole item (box inside box)
a = [1,2,3,4]
a.append([5,6])
print(a)

# extend : add elements seperately (open box and mix)
a.extend([5,6])
print(a)
```

```python
# insert (index,value)
a = [1,2,3,4]
a.insert(1,100)
print(a)
```

```python
# index & count 
a = [2,2,3,4,2]
print(a.index(2)) # first position
print(a.count(2)) # how many times repeat 

```
