```python
import numpy as np 
import random 
```

```python
# Method Overriding 

class SavingAccount:
    def __init__(self):
        self.number_of_cupons = 0 

    def getCupons(self):
        self.number_of_cupons += np.random.randint(3, 6)

class BusinessAccount(SavingAccount):
    
    def getCupons(self):
        self.number_of_cupons += np.random.randint(10, 16) 
```

```python
p1 = SavingAccount() 
p2 = BusinessAccount() 
```

```python
p1.number_of_cupons
```

```python
p1.getCupons() 
```

```python
p1.number_of_cupons
```

```python
p2.number_of_cupons
```

```python
p2.getCupons() 
```

```python
p2.number_of_cupons
```

```python
class Number:
    def __init__(self, n):
        self.n = n 
    
    def __add__(self, other):
        return Number(self.n + other.n) 

    def __sub__(self, other):
        return self.n - other.n 

    def __mul__(self, other):
        return self.n * other.n 
    
    def __gt__(self, other):
        return self.n > other.n 
    
    def __ge__(self, other):
        return self.n >= other.n 

    def __str__(self):
        # return f"Data : {self.n}\nMemory Location : {id(self.n)}\nData  Type : {type(self.n)}"  
        return f"{self.n}"
```

```python
a1 = Number(10) 
b1 = Number(20) 
```

```python
print(a1) 
```

```python
a1.n
```

```python
b1.n
```

```python
a1 + b1 # def __add__(self, other) (self --> current object : a1, other --> other object : b1)
```

```python
a1.__add__(b1) 
```

```python
a1.n + b1.n 
```
