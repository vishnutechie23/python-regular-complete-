```python
# Accessing (Indexing : Forward & Negative)

tup = (8,4,2,3,1,9,23,40)
print(f"Original Tuple : {tup}") 

print(f"First Element : {tup[0]}")
print(f"Last Element : {tup[-1]}") 
```

### DEfault values of slicer : (start:0, Stop:len(tuple), Step:1)

```python
# Slicing (slicer[start:stop:step])

print(f"First 3 Element : {tup[0:3:1]}") 
print(f"Last 3 Element : {tup[-1:-4:-1]}")
print(f"Length : {tup[0:len(tup):1]}")
print(f"Length : {tup[:len(tup):1]}")
print(f"Length : {tup[:len(tup):]}") 

```

## Concept of mutability : (Tuple : immutable)
### Insertion, updation, Deletion Not allowed 


```python
# Looping (All three types of for loop )
# Index based for loop (Index --> Accessin --> Data)
for i in range(len(tup)):
    print(f"Index :{i}, Data : {tup[i]}")
```

```python
# item baseds for loop 
for item in tup:
    print(item,end=" ")
```

```python
# Enumerate based for loop (Index as well as Data)

for index,item in enumerate(tup):
    print(F"Index : {index}, Data : {item}")
```

```python
# Membership 
print(f"{40} in {tup} : {40 in tup}")
print(f"{100} in {tup} : {100 in tup}")
print(f"{0} in {tup} : {0 not in tup}")
print(f"{2} in {tup} : {2 not in tup}") 
  
```

```python
# passing tuple as parameter 
from typing import Tuple 
```

```python
def add(a:int, b :int , *x : Tuple[int]) -> int:
    print(f"a = {a}\nb = {b}\nx = {x}")
    ans = a + b 
    for item in x:
        ans += item
    return ans 
```

```python
add(1,2)
```

```python
add(1,2,3)
```

```python
add(1,2,3,4)
```

```python
def add(a :int , b:int, *x : Tuple[int]) -> int:
    return a + b + sum(x)
```

```python
add(1,2,3,4,5)
```

```python
# Tuple comprehension 
t = (item for item in input().split()) # Tuple[str] 
print(t)
li = list(t) 
print(li) # List of str 
```
