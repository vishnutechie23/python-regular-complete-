```python
import math 
# ax2 + bx + c = 0
a = float(input("Enter coefficient a :"))
b = float(input("Enter coefficient b :"))
c = float(input("Enter coefficient c :")) 
# check wheather real or not 
check = b ** 2 - 4*a*c 
# condition if b**2 - 4*a*c is positive,zero or negetive 
if check > 0 : #two real roots 
    root1 = ( -b + math.sqrt(check)) / (2 * a)
    root2 = ( -b + math.sqrt(check)) / (2 * a) 
    print(f"Root 1 : {root1}")
    print(f"Root 1 : {root2}") 
elif check == 0: # one real root
    root = -b / (2 * a)
    print(f"Root : {root}")
else :
    # not real / complex root 
    real_check  = -b / (2 *a)
    notreal_part = math.sqrt(abs(check) / (2 *a))
    print(f"Root 1 : {real_check} + {notreal_part}i")
    print(f"Root 2 : {real_check} + {notreal_part}i")
```
