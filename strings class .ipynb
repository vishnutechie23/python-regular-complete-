```python
s1, s2, s3 = "123", "abc", "abc123" 

print(f"String : {s1}\nNumeric : {s1.isnumeric()}\nAlphabetic : {s1.isalpha()}\nAlpha-Numeric : {s1.isalnum()}\n")
print(f"String : {s2}\nNumeric : {s2.isnumeric()}\nAlphabetic : {s2.isalpha()}\nAlpha-Numeric : {s2.isalnum()}\n")
print(f"String : {s3}\nNumeric : {s3.isnumeric()}\nAlphabetic : {s3.isalpha()}\nAlpha-Numeric : {s3.isalnum()}\n")
```

```python
new_string = "abc123!@#"

print(f"String : {new_string}\nNumeric : {new_string.isnumeric()}\nAlphabetic : {new_string.isalpha()}\nAlpha-Numeric : {new_string.isalnum()}")
```

```python
# given a string, count number of alphabets, digits and special characters 

aura = "abc123!@#  " 

a, d = 0, 0 

for char in aura:
    if char.isalnum():
        if char.isnumeric():
            d += 1  
        else:
            a += 1 

    
print(f"String : {aura}\nAlphabets : {a}\nDigits : {d}\nSpecial Characters : {len(aura) - a - d}")
```

```python
def isPalindrome(n : int) -> bool:
    n = str(n) 
    return n == n[::-1]
```
