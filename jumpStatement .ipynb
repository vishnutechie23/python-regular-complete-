# Jump Statements

### i. Break (Stop the iteration)
### ii. Continue (Skip the iteration)

```python
a = 1

while a < 11:
    if a == 5:
        break
    print(a , end=" ")
    a += 1
```

```python
a = 1

while a < 11:
    print(a , end=" ")
    if a == 5:
        break
    a += 1
```

```python
a = 1

while a < 11:
    if a % 2 == 1:
        a += 1
        continue
    else:
        print(a, end=" ")
        a += 1
```
