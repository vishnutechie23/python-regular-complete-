```python
# Accessing (Indexing : Forward & Negative) 

t = (8, 2, 4, 1, 5, 10, 20, 40)
print(f"Original Tuple : {t}") 

print(f"First Element : {t[0]}")
print(f"Last Element : {t[-1]}")

# Slicing (Slicer[Start : Stop : Step])

print(f"First 3 Elements : {t[0:3:1]}")

# Default Values of Slicer : (Start : 0, Stop : len(tuple), Step : 1)
print(f"First 3 Elements : {t[:3:]}")
```

```python
# Looping (All the three types of for loops) 

# Index based for loop (Index --> Accessing --> Data)
for i in range(len(t)):
    print(f"Index : {i}, Data : {t[i]}")
```

```python
# Item Based for loop 

for item in t:
    print(item) 
```

```python
# Enumerate Based (Index as well as Data) 

for index, item in enumerate(t):
    print(f"Index : {index}, Data : {item}")
```

```python
# Membership

print(f"{40} in {t} : {40 in t}")
print(f"{40} not in {t} : {40 not in t}")
```

```python
from typing import Tuple
```

```python
def add(a : int, b : int, *x : Tuple[int]) -> int:
    print(f"a = {a}\nb = {b}\nx = {x}")
    ans = a + b 

    for item in x:
        ans += item 
    
    return ans 
```

```python
add(1, 2) 
```

```python
add(1, 2, 3) 
```

```python
add(1, 2, 3, 4) 
```

```python
def add(a : int, b : int, *x : Tuple[int]) -> int:
    return a + b + sum(x) 
```

```python
add(1, 2, 3,  4, 5) 
```

```python
# Tuple Comprehension 

t = (item for item in input().split()) # Tuple[str] 

print(t) 

li = list(t) 

print(li) 
```
