```python
# permtation and combination 
n = int(input()) 
r = int(input()) 
n_fact,r_fact,n_r_fact = 1,1,1
# n_fact = r_fact = n_r_fact = 1 
for i in range (1 , n+1):
    n_fact *= i 
for i in range (1 , r+1):
    r_fact *= i 
for i in range (1 , n - r +1):
    n_r_fact *= i

answer = n_fact / (r_fact * n_r_fact)
print(f"n = {n} and r = {r}")
print(f"{n}C{r} = {answer}")
```

```python
def factorial(n : int) -> int:
    ans = 1 

    for i in range(2, n+1):
        ans *= i 
    
    return ans 
```

```python
n = int(input("n = "))
r = int(input('r = '))

ans = factorial(n) / (factorial(r) * factorial(n-r))

print(f"{n}C{r} = {ans}")
```

```python
# Prime Number 

def isPrime(n : int) -> bool:
    if n < 2:
        return False 
    
    for i in range(2, n):
        if n % i == 0:
            return False 
    else:
        return True 
```

```python
for i in range(1, 101, 1):
    print(f"{i} --> {'Prime' if isPrime(i) == True else 'Not Prime'}",end= " ")
```
