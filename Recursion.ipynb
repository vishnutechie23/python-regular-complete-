```python
def s(n : int) -> int | str:
    # Edge Case 
    if n < 1:
        return "Invalid Input"
    # Base Case 
    if n == 1:
        return 1
    # Induction Hypothesis 
    small_ans = s(n-1) # My Assumption that this is already calculated 
    # Induction Step 
    final_ans = small_ans + n 
    return final_ans
```

```python
n = int(input("Enter the number : "))

ans = s(n) 

print(f"Sum of first {n} Natural Numbers is {ans}")
```

```python
def s2(n : int) -> int:
    # Base Case 
    if n == 1:
        return 1
    # Induction Hypothesis 
    print(f"Function Call For {n}")
    small_ans = s2(n-1) # My Assumption that this is already calculated 
    print(f"Answer for s({n-1}) = {small_ans}")
    # Induction Step 
    final_ans = small_ans + n 
    print(f"Answer for final ans for s({n}) = s{n-1} + {n} i.e. {small_ans} + {n} --> {final_ans}")
    return final_ans
```

```python
def sum_iterative(n : int) -> int:
    ans = 0 
    for i in range(1, n+1):
        ans += i 
    
    return ans 
```

```python
def fact_iterative(n : int) ->  int:
    ans = 1 
    for i in range(2, n+1):
        ans *= i 
    
    return ans 
```

```python
def fact_recursive(n : int) -> int:
    # Base Case 
    if n == 0:
        print(f"Base Case : n = {0} --> {0}! = 1")
        return 1 
    if n == 1:
        print(f"Base Case : n = {1} --> {1}! = 1")
        return 1 
    # Induction Hypothesis 
    print(f"Current Function Call : {n}")
    small_ans = fact_recursive(n-1) 
    print(f"{n-1}! = {small_ans}")
    # Induction Step 
    final_ans = small_ans * n
    print(f"{n}! = {n-1}! * {n} --> {small_ans} * {n} --> {final_ans}") 
    return final_ans
```

```python
fact_recursive(5) 
```

```python
def fact_optimised(n : int) -> int:
    # Base Case 
    if n == 0 or n == 1:
        return 1 
    # Induction Hypothesis & Induction Step 
    return fact_optimised(n-1) * n
```

```python
def fibonacci_iterative(n : int) -> int | str:
    # Edge Case 
    if n < 0:
        return "Invalid Input"

    if n == 0 or n == 1:
        return n 
    
    n1, n2 = 0, 1
    for i in range(n-1):
        n1, n2 = n2, n1 + n2 
    
    return  n2
```

```python
for i in range(0, 11, 1):
    print(F"{i} --> {fibonacci_iterative(i)}")
```

```python
def fibo_recursive(n : int) -> int | str:
    # Edge Case 
    if n < 0:
        return "Invalid Input"
    
    # Base Case 
    if n == 0 or n == 1:
        return n 

    # Induction Hypothesis 
    small_ans1 = fibo_recursive(n-1) 
    small_ans2 = fibo_recursive(n-2) 

    # Induction Step 
    final_ans = small_ans1 + small_ans2
    return final_ans 
```

```python
for i in range(0, 11, 1):
    print(f"{i} --> {fibo_recursive(i)}")
```

```python
def fibo_recursive_optimised(n : int) -> int | str:
    # Edge Case 
    if n < 0:
        return "Invalid Input"
    
    # Base Case 
    if n == 0 or n == 1:
        return n 

    # Induction Hypothesis & Induction Step 
    return fibo_recursive(n-1) + fibo_recursive(n-2) 
```

```python
fibo_recursive_optimised(10) 
```

# Recursion In  List / Array 

```python
from typing import List 
```

```python
# I will check for the first element, then recursion will handle the rest 
# Check for the first element, and then call the recursion for rather part of the list
    
def isPresent(li : List[int], n : int) -> bool:
    # Base Case 
    if len(li) == 0:
        return False 

    if li[0] == n:
        return True 
    # Induction Hypothesis 
    small_ans = isPresent(li[1:], n) 
    # Induction Step 
    return small_ans 
```

```python
def isPresentBetter(li : List[int], n : int, index = 0) -> bool:
    if index == len(li):
        return False 
    
    if li[index] == n:
        return True 
    
    small_ans = isPresentBetter(li, n, index + 1) 

    return small_ans 
```

```python
li = [int(x) for x in input("Enter the list elements : ").split()]
n = int(input("Enter the number to check : "))

print(f'{n} Present In {li} : {isPresentBetter(li, n)}')
```

```python
def firstIndex(li : List[int], n : int, index = 0) -> int:
    if index == len(li):
        return -1 
    
    if li[index] == n:
        return index 
    
    small_ans = firstIndex(li, n, index + 1) 
    return small_ans
```

```python
li = [int(x) for x in input().split()] 
n = int(input()) 

index = firstIndex(li, n) 

if index == -1:
    print(f"{n} not found in {li}")
else:
    print(f"{n} found in {li} at index {index}")
```

```python
# Last Index 

def lastIndex(li : List[int], n : int, index = 0) -> int:
    if len(li) == index:
        return -1
    
    ans = lastIndex(li, n, index + 1) 

    if ans == -1:        
        if li[index] == n:
            return index 
        else:
            return -1 
    else:
        return ans 
```

```python
li = [int(x) for x in input().split()] 
n = int(input()) 

index = lastIndex(li, n) 

if index == -1:
    print(f"{n} not found in {li}")
else:
    print(f"{n} found in {li} at index {index}")
```

```python
def allIndex(li : List[int], n = int, index = 0, index_list = []) -> List[int]:
    if len(li) == index:
        return index_list 
    
    if li[index] == n:
        index_list.append(index) 
    
    return allIndex(li, n , index + 1, index_list) 

li = [int(x) for x in input().split()] 
n = int(input()) 

index = allIndex(li, n) 

print(f"{n} found in {li} at these indexes {index}")
```
