To calculate simple intrest

```python

p = 100000
r = 10
t = 60 
si = p * r * t / 100 * 12
```

```python
#1. normal method
print("The simple Interest for" ,p,  "at" ,r, "% rate of interest for" ,t, "months is" ,si ) 
```

```python
# f-string mwthod (formatting string literal)
print(f"The simple Interest for {p}  at {r} % rate of interest for {t} moths is {si} ") 
```

```python
#3. Format method 
print( "The simple Interest for {} at {} % rate of interest for {} months is {}" .format (p,r,t,si)) 
```

For Area of Triangle

```python
base = 100
height = 50 
AREA = 0.5 * base * height
print(f"The area of Triangle is {AREA}")
```

For Area of Rectangle

```python
l = 100
b = 50
area = l * b 
print(f"The area of rectangle is {area}")
```

For Area of Square

```python
side = 10 
area = side * side 
print(f"The area of Square is {area}")
```

For Area of Circle

```python
R = 15
pi = 3.14
area = pi * R * R
print(f"The area of Circle is {area}")
```
