# Accessing 

```python
string = "Orewa Namikaze Minato"

print(f"Original String : {string}")

print(f"First Character : {string[0]}")
print(f"Last Character : {string[-1]}")
print(f"{string[-6]}")
```

# Looping 

1. Index 
2. Item 
3. Enumerate 

```python
print(f"Original String : {string}")
```

```python
# Index Based --> range(start, stop, step) 

for index in range(0, len(string), 1):
    print(f"Index : {index}, Data : {string[index]}, ASCII Value : {ord(string[index])}")

```

```python
for character in string:
    print(character, end = "\n") 
```

```python
for index, item in enumerate(string):
    print(f"Index : {index}, Data : {item}")
```

```python
# Functionalities in String 
```

```python
# Concatenation & Repetition 

s1 = "Uzimaki" 
s2 = "Naruto" 

final_string = s1 + " " + s2 
custom_string = (s2 + " ") *  2 

print(f"String 1 : {s1}\nData Type : {type(s1)}\nMemory Location : {id(s1)}\n")
print(f"String 2 : {s2}\nData Type : {type(s2)}\nMemory Location : {id(s2)}\n")
print(f"final string : {final_string}\nData Type : {type(final_string)}\nMemory Location : {id(final_string)}\n")
print(f"Custom String : {custom_string}\nData Type : {type(custom_string)}\nMemory Location : {id(custom_string)}\n")
```

```python
# Split & Join Function 
```

```python
string = "It's best not to be optimistic and think it's an illusion. The pain is not an illusion. This pain is absolutely no different from reality. How long will your spirit sustain?"
```

```python
# Character by Character 
for item in string:
    print(item) 
```

```python
# Word by Word 
for item in string.split():
    print(item) 
```

```python
li = string.split() 
print(li) 
```

```python
new_string = " ".join(li) 
print(new_string)
```
