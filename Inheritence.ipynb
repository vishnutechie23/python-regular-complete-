```python
class Products:
    dmart_products = {
        "Aashirvaad Atta 5kg": 275,
        "India Gate Basmati Rice 1kg": 190,
        "Fortune Sunflower Oil 1L": 155,
        "Amul Butter 500g": 285,
        "Tata Salt 1kg": 28,
        "Surf Excel Easy Wash 1kg": 120,
        "Colgate Strong Teeth Toothpaste 200g": 110,
        "Dove Soap 100g (Pack of 3)": 165,
        "Maggi Noodles Pack of 12": 168,
        "Parle-G Biscuits 800g": 95,
        "Good Day Cashew Cookies 1kg": 180,
        "Tata Tea Premium 1kg": 510,

        # Products less than 500g
        "Amul Butter 100g": 60,
        "Britannia Good Day 200g": 40,
        "Tata Tea Premium 250g": 140,
        "Surf Excel Detergent 500g": 65,
        "Colgate Toothpaste 100g": 55,
        "Maggi Noodles Single Pack 70g": 14,
        "Dairy Milk Chocolate 45g": 20,
        "Lay's Classic Chips 52g": 20
    }
    
    def __init__(self, name, contact):
        self.name = name 
        self.contact = contact 
        self.discounted_price = self.price = 0
        self.discount = np.random.randint(1, 6)
        self.cart = [] 

    def add_to_cart(self, product_name):
        self.cart.append(product_name)

    def applyDiscount(self):
        self.discounted_price = self.price * (100 - self.discount) / 100   

    def purchase(self):
        for item in self.cart:
            self.price += Products.dmart_products[item]

    def displayBill(self):
        print(f"Name : {self.name}\nContact : {self.contact}") 

        for item in self.cart:
            print(F"{item} - {Products.dmart_products[item]}")
        
        print(f"Discount : {self.discount}%")

        print(f"Final Ammount : {self.discounted_price}") 

        print(f"You are Saving {self.price - self.discounted_price : 0.2f} Rupees")
```

# Hybrid Inheritence 

```python
import numpy as np 
import random 
```

```python
class BankAccount:
    min_account_balance = 5000 
    bank_name = "HDFC Bank"
    IFSC_code = "HDFC" + "".join([str(item) for item in np.random.randint(0, 10, 6)]) 

    def __init__(self, name, contact, balance, email, address, age):
        self.name = name 
        self.contact = contact 
        self.balance = balance 
        self.age = age 
        self.email = email 
        self.address = address 
        self.account_number = "".join([str(item) for item in  np.random.randint(0, 10, 15)])

    def deposit(self, ammount):
        self.balance += ammount 
        print(f"Rupees {ammount}/- Deposited in {self.account_number} Successfully.")

    def withdraw(self, ammount):
        if self.balance - ammount < BankAccount.min_account_balance:
            print(f"Insufficient Funds.")
        else:
            self.balance -= ammount 
            print(f"{ammount} Debited from {self.account_number} Successfully\nAvl Bal : {self.balance}") 

    def displayDetails(self):
        print(f"Bank : {BankAccount.bank_name}\nAccount Number : {self.account_number}\nIFSC Code : {BankAccount.IFSC_code}\nName : {self.name}\nEmail : {self.email}\nContact Number : {self.contact}\nAge :  {self.age}\nAddress : {self.address}\nBank Balance :  {self.balance}")
```

```python
class FDAccount(BankAccount):
    fine_percentage = 10 
    def __init__(self, name, contact, balance, email, address, age, fd_amount, fd_rate, time_period):
        super().__init__(name, contact, balance, email, address, age)
        self.fd_amount = fd_amount 
        self.fd_rate = fd_rate
        self.time_period = time_period
        self.isFdMatures = False 
    
    def fd_matured(self):
        self.isFdMatures = True 
    
    def fd_break(self):
        if self.isFdMatures:
            self.balance = self.balance + (self.fd_amount * (1 + (self.time_period * self.fd_rate * 0.01)))
        else:
            self.balance += (self.fd_amount * (100 - FDAccount.fine_percentage) * 0.01) 

        self.fd_amount = 0 
    
    # method overriding 
    def displayDetails(self):
        super().displayDetails()
        print(f"Fd Ammount : {self.fd_amount}\nInterest Rate : {self.fd_rate}\nTime Period : {self.time_period}")
```

```python
class CurrentAccount(BankAccount):

    due_interest = 12 
    def __init__(self, name, contact, balance, email, address, age):
        super().__init__(name, contact, balance, email, address, age) 
        self.credit_limit = ((self.balance // 1000) * 1000) * 20 
        self.due = 0 
        self.interest_added = False 

    def getCreditAmmount(self, amount):
        self.balance += amount 
        self.due += amount 
        print(f"{amount} Credited to {self.account_number} Successfully.") 
    
    def repayAmount(self, amount):
        if amount > self.due:
            self.balance += amount - self.due
            self.due = 0  
        else:
            self.due -= amount 
        
        if self.due == 0:
            self.interest_added = False 

    def calculate_final_due(self):
        if self.interest_added:
            return 
        self.due = self.due * ((100 + CurrentAccount.due_interest) / 100) 
        self.interest_added = True 
```

```python
class StudentAccount(FDAccount, CurrentAccount):
    fd_interest = 10
    due_interest = 8

    def __init__(self, name, contact, balance, email, address, age, fd_amount = 0 , fd_rate = 0, time_period = 0):
        FDAccount.__init__(self, name, contact, balance, email, address, age, fd_amount, fd_rate, time_period) 
        CurrentAccount.__init__(self, name, contact, balance, email, address, age) 
        self.cupons = np.random.randint(1, 4) 

    def calculate_final_due(self):
        if self.interest_added:
            return 
        self.due = self.due * ((100 + StudentAccount.due_interest) / 100) 
        self.interest_added = True 

    def getFdBonus(self):
        self.balance = self.balance + (self.fd_amount * (1 + (self.time_period * self.fd_rate * 0.01)))

        self.fd_amount = 0 

    def getStudentCupons(self):
        self.cupons += np.random.randint(1, 5)      

```
