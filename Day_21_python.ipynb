## List slicing  

```python
li = [1,3,4,23,43,55,648]
print(li) 
```

```python
print(f"Original List : {li}\n\nForward Indexing")

print(f"Frist 3 Elements : {li[0:3:1]}")
print(f"Last 3 Elements : {li[4:7:1]}")
print(f"Every Element Except First and Last : {li[1:6:1]}") 
```

```python
print(f"Original List : {li}\n\nNegative Indexing")

print(f"Frist 3 Elements : {li[-7:-4:1]}")
# print(f"Last 3 Elements : {li[]}")
# print(f"Every Element Except First and Last : {li[1:6:1]}") 
```

```python
print(li)
print(f"{li[-1:-4:1]}") 
```

```python
print(li)
print(li[-3:-1:1] + [li[-1]]) 
```
