## List slicing  

```python
li = [1,3,4,23,43,55,648]
print(li) 
```

```python
print(f"Original List : {li}\n\nForward Indexing")

print(f"Frist 3 Elements : {li[0:3:1]}")
print(f"Last 3 Elements : {li[4:7:1]}")
print(f"Every Element Except First and Last : {li[1:6:1]}") 
```

```python
print(f"Original List : {li}\n\nNegative Indexing")

print(f"Frist 3 Elements : {li[-7:-4:1]}")
# print(f"Last 3 Elements : {li[]}")
# print(f"Every Element Except First and Last : {li[1:6:1]}") 
```

```python
print(li)
print(f"{li[-1:-4:1]}") 
```

```python
print(li)
print(li[-3:-1:1] + [li[-1]]) 
```

```python
print(li)
li[5:1:-1]
```

```python
li[2:6:1]
```

```python
li[-2:-6:-2]
```

# Default values in the slicer/ slicing 
## start = 0 , stop = len(list) , step = 1 

```python
li = [8,9,2,4,5,7,1]
```

```python
# For full list 
print(li[0:7:1])
print(li[:7:1])
print(li[::1])
print(li[::])
print(li[:])
```

```python
print(li[0:4:1])
print(li[0:4:])
print(li[:4:]) # Most Prefered (Good practice)
print(li[:4])
print(li[:]) #Full list 
# print(li[4:]) # problem 
```

```python
print(li[3:7:1])
print(li[3:7:])
print(li[3::])
print(li[3:]) # Not prefered 
```

```python
# Reverse a list 

print(li) 
print(li[::-1])
print(li[::-2])
print(li[::-3])
```

```python
# check Memory location becuz -5 se 256 tak same dabba uske upar niche change location 
a = 1000 
b = 1000 

print(f"a = {a}, id = {id(a)}")
print(f"b = {b}, id = {id(b)}")
```
