```python
class AppleStore:
    
    product_price = {
        "i phone 12 pro max": 600000,
        "i phone 13 pro max": 700000,
        "i phone 14 pro max": 800000,
        "i phone 15 pro max": 100000,
        "i phone 16 pro max": 120000,
        "i phone 17 pro max": 150000,
    }

    def __init__(self, name, email, contact_no, productName):
        self.name = name
        self.email = email
        self.contact_no = contact_no
        self.productName = productName
        
        # User jo dale uska price nikalna
        self.price = self.product_price[self.productName]

    def display(self):
        print("Customer:", self.name)
        print("Product:", self.productName)
        print("Price:", self.price)
```

```python
vishnu = AppleStore(
    "vishnu",
    "vishnu@123",
    "1234567890",
    "i phone 16 pro max"
)

vishnu.display()
```
