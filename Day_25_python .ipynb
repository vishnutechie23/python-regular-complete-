```python
# n = int(input("Enter a number"))
n = 5
for i in range(n):
    for sp in range(n - i - 1):
        print(" ",end=" ")
    for st in range(i+3):
        print("*",end=" ")
    for st in range(i+2):
        print("*",end=" ")
    print()

for i in range(n):
    for st in range(2*n+3):
        print("*",end=" ")
    print()

for i in range(n):
    for sp in range(i):
        print(" ",end=" ")
    for st in range(n-i +4):
        print("*",end=" ")
    for st in range(n-i-1):
        print("*",end=" ")
    print()


```
