```python
n = int(input("Enter a number"))
start = 1 
T = 0
while start < n + 1:
    T = T + start
    start += 1
    print(T,end="") 
```

```python
n = int(input("Enter a number"))
start = 1 
T = 0
while start < n + 1:
    if T == T + start:
        print(T)
        start += 1
    else:
        start +=1 
```
