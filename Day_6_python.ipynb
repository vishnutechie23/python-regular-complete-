## Logical graphs and their coding perspective

```python
def logical_graph(a, b):
    # part1 = a and b
    # part2 = not a or b 
    # part3 = part1 or b 
    # part4 = not part3 and part2
    # return part4
    return  (not ((a and b) or b ) and (not a or b))
```

```python
def logical_graph(a, b, c):
    # part1 = a or b
    # part2 = a or c
    # part3 = part1 and part2 
    # part4 = part3 or not part2 
    # part5 = not part3 and part4 
    # return part5
        return not ((a or b) and (a or c)) and (((a or b) and (a or c)) or not (a or c)) 
```
