## Jump statements 
#### break --> stop the Iteration 

```python
#print upto 5 without change in code
a = 1 
while a <= 10:
    print(a,end=" ")
    if a == 5:
        break
    
    a +=1
```

```python
st = int(input("Enter a number"))
ref = st
stop = 0
fact = 1 
while st > stop:
    fact = fact * st
    st -=1
print(f"{ref}! is {fact}")
```

#### continue --> Skip the iteration 

```python
a = 1

while a < 11:
    if a % 2 == 1:
        a += 1
        continue
    else:
        print(a, end=" ")
        a += 1
```

## Infinte loop 

# Syntax

# while True:
#     Logic

```python
a = 2
while True:
    if a < 11 :
        print(a)
        a += 2
    else:
        break
```

```python
# print number from 1 to 10 using infinte loop

a = 1
while True:
    if a > 10:
        break
    print(a,end=" ")
    a += 1
```

```python

a = 1 
while True:
    if a > 10 :
        break
    if a % 2 == 0:
        print(a)
    a += 1
    
```

```python
# print  all even numbers from 1 to 10 using infinte loop 
a = 1 
while True:
    if a > 10:
        break
    elif a % 2 == 0:
        print(a)
        a += 1
    else:
        a += 1
        continue 
```

```python

a = 1 
while True:
    if a > 10:
        break
    if a % 2 == 1:
        a += 1
        continue
    print(a,end=" ")
    a += 1 
```

```python
#  Better Version (Low Complexity)

# Print all even numbers from 1 to 10 using infinite loop

a = 1

while True:
    if a > 10:
        break
    elif a % 2 == 0:
        print(a, end=" ")
        a += 1
    else:
        a += 1
        continue
```
