## Map (function,iterable)
### Apply a particular function on each element of the iterable  

```python
import numpy as np
import random 
```

```python
li = [int(x) for x in np.random.randint(1,10,10)] 
print(li) 
```

```python
squared_list = list(map(lambda x : x ** 2, li)) 
print(f"Original list : {li}\nSquared List : {squared_list}") 
```

```python
odd_even_bool = list(map(lambda x : True if  x % 2 == 0 else False, li ))
```

```python
print(li)
print(odd_even_bool)
```

```python
li = [int(x) for x in np.random.randint(1,10,10)]
odd_even_bool = list(map(lambda x : True if x % 2 == 0 else False,li)) 
even_list = []
odd_list = []
for index, item in enumerate(odd_even_bool):
    if item:
        even_list.append(li[index])
    else:
        odd_list.append(li[index])

print(li)
print(odd_even_bool)
print(even_list)
print(odd_list)
```

```python
for item in odd_even_bool:
    print(item)
```

## Filter(function, iterable)
#### filteres the data based on a particular condition 

```python
li = [int(x) for x in range(1,11,1)]
print(li)
```

```python
odd_list = list(filter(lambda x : x % 2 != 0,li))
print(odd_list)
```

```python
first_name = ["zahoor","sunny"]
last_name = ["mistry", "kazi"]

full_name = list(map(lambda x , y : x + " " +y,first_name,last_name))
print(full_name)
```

```python
li = ["python","java","c++","javaScript"] 
new_list = list(filter(lambda x : len(x) >= 5,li))
print(f"Original List : {li}")
print(f"Updated List : {new_list}")
```

## reduce(function, iteration)
#### Returns a single value

```python
from functools import reduce 
```

```python
li = [int(x) for x in range(1,6,1)]
print(li)
```

```python
ans = 0 
for item in li:
    ans += item 

print(f"The sum of all element in {li} is {ans}")
```

```python
solution = reduce(lambda x , y : x + y, li,0) 
print(solution)
```

```python
print(f"Sum of all the element in the list ")

solution = reduce(lambda x , y : x+y , list(map(int, input().split())),0)
print(f"The sum is {solution}")
```

```python
solution = reduce(lambda x , y : x * y,list(map(int,input().split())),1) 
print(solution)
```

```python
#Fibonacci 
li = [(0, 0), (1, 1)] 

n = int(input()) 
for item in range(n-1):
    print(f"Original List : {li}\nNew Element Calculated : {li[item][1]} + {li[item + 1][1]} = {li[item][1] + li[item + 1][1]}\nFinal Element : {(item + 2, li[item][1] + li[item + 1][1])}\n")
    li.append((item + 2, li[item][1] + li[item + 1][1])) 

print(li) 
```

```python
naruto = ["Orewa", "Namikaze", "Minato", "Shano", "Sharingan", "Ameterasu", "Mangekyou", "Sharingan", "Byakugan"] 

reduce(lambda x, y : x + " " + y, naruto) 
```
