```python
n = 9

for i in range(n):
    for j in range(n):
        # (i = 0, j = 0) (i = 0, j = n-1) (i = n -1, j = 0) (i = n-1, j = n-1)
        if (i == 0 and j == 0) or (i == 0 and j == n-1) or (i == n-1 and j == 0) or (i == n-1 and j == n-1):
            print("üòé", end = " ") 
        # row - 0 or row n-1 --> - 
        elif i == 0 or i == n-1:
            print("-", end = " ")
        # # col = 0 or col = n-1 --> |
        elif j == 0 or j == n-1:
            print("ü´°", end = " ")  
        else:
            print(" ", end = " ") 
    print() 
```

```python
n = 9

for i in range(n):

    # V
    for j in range(n):
        if (j == i and i <= n//2) or (j == n-i-1 and i <= n//2):
            print("1", end=" ")
        else:
            print(" ", end=" ")
    print(" ", end="")

    # I
    for j in range(n):
        if i == 0 or i == n-1 or j == n//2:
            print("1", end=" ")
        else:
            print(" ", end=" ")
    print("  ", end="")

    # S
    for j in range(n):
        if i == 0 or i == n//2 or i == n-1 or (j == 0 and i < n//2) or (j == n-1 and i > n//2):
            print("1", end=" ")
        else:
            print(" ", end=" ")
    print("  ", end="")

    # H
    for j in range(n):
        if j == 0 or j == n-1 or i == n//2:
            print("1", end=" ")
        else:
            print(" ", end=" ")
    print("  ", end="")

    # N
    for j in range(n):
        if j == 0 or j == n-1 or j == i:
            print("1", end=" ")
        else:
            print(" ", end=" ")
    print("  ", end="")

    # U
    for j in range(n):
        if j == 0 or j == n-1 or i == n-1:
            print("1", end=" ")
        else:
            print(" ", end=" ")

    print()
```
