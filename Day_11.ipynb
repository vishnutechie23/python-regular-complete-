## Jump Statements  

Break --> Stop  the Iteration 

```python
a = 1 

while a <= 10:
    if a == 5:
        break
    print(a, end = " ") 
    a += 1 
```

```python
a = 1 

while a <= 10:
    print(a, end = " ") 
    if a == 5:
        break
    a += 1 
```

Continue --> Skip the Iteration 

```python
# This is an infinite loop --> skip operation performed before updation 
'''
a = 1 

while a < 11:
    if a % 2 == 1:
        continue 
    else:
        print(a, end = " ")
    a += 1 
'''
```

```python
a = 1 

while a < 11:
    if a % 2 == 1:
        a += 1
        continue 
    else:
        print(a, end = " ")
        a += 1 
```

```python
a = 1 
while a < 11:
    if a % 2 == 0:
        print(a, end = " ")
    a += 1 
```

```python
# print numbers from 1 to 10 using an infinite loop 

a = 1 

while True:
    if a > 10:
        break
    print(a, end = " ") 
    a += 1
```

```python
# print all even  numebers from 1 to 10  using infinite loop 

a = 1 

while True:
    if a > 10:
        break 
    elif a % 2 == 0:
        print(a, end = " ") 
        a += 1 
    else:
        a += 1 
        continue 
```

```python
# Prime Number Detector 


n = int(input()) 

if n < 2:
    print(f"{n} is not a Prime Number ")
else:
    for i in range(2, n):
        if n % i == 0:
            print(f"{n} is not a prime number")
            break 
    else:
        print(f"{n} is a prime number") 
```

```python
# Nested Looping 

for outer_loop in range(1, 6):
    print(f"{outer_loop}", end = " --> ")
    for inner_loop in range(11, 14):
        print(f"{inner_loop}", end = " ")
    print() 
```
