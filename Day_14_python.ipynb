## Diamond  pattern

```python
n = 4 
for i in range(n):
    for st1 in range(i+1):
        print("*",end=" ")
    for sp in range(2 *(n - i -1)):
        print(" ",end=" ")
    for st2 in range(i+1):
        print("*",end=" ")   
    print()     
```

```python
n = int(input("Enter a numeber"))
for i in range(n):
    for j in range(i +1):
        print("*",end=" ")
 
    if i == n-1:
        for sp in range(2*(n-i-1)):
            print(" ",end=" ")
        for st in range(i):
            print("*",end=" ")
    else:
        for sp in range(2*(n-i-1)-1):
            print(" ",end=" ")
        for st in range(i+1):
            print("*",end=" ")
    print()        
```

```python
n = int(input("Enter a number"))
for i in range(n):
    for sp in range(n - i -1):
        print(" ",end=" ")
    for st in range(2 * i +1):
        print("*",end=" ")
    print()

for j in range(n-1):
    for sp in range(j+1):
        print(" ",end=" ")

    for st in range(2*(n - j -1)-1):
        print("*",end=" ")
    print()
```

```python
r = 7  # radius

for i in range(-r, r + 1):
    for j in range(-r, r + 1):
        if i*i + j*j <= r*r:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()


```
