```python
# QS 11 : count how many even and odd numbers between 1 to 20 
st = 1 
stop = 20
count_even = 0
count_odd = 0
while st <= stop:
    if st % 2 == 0:
        count_even += 1
    else:
        count_odd += 1
    st+=1

print(count_even)
print(count_odd)
```

```python
n = 121
temp = n
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10

if rev == temp:
    print("Palindrome")
else:
    print("Not Palindrome")
```
