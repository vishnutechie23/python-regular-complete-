## Operators : identity

```python
a = 10
b = 10
print(f"{a} == {b} : {a == b }")
print(f"{a} IS {b} : {a is b }") 
```

```python
a = 10
b = 20
print(f"{a} == {b} : {a == b }")
print(f"{a} IS {b} : {a is b }") 
```

```python
a = 1000
b = 1000


print(f"{a} == {b} : {a == b }")
print(f"{a} IS {b} : {a is b }")
print(f"Memeory Location of {a} : {id(a)}\nMemory Location of {b} : {id(b)}") 
```

## Operators : membership Operator 


```python
li = [8, 10, 2 ,30, 4]
print(f"List : {li}")
print(f"8 in {li} : {8 in li}")
print(f"10 in {li} : {10 in li}")
print(f"100 in {li} : {100 in li}")
```

## Operators : unary 
"+ - ~ not "

```python
a = 10 
print(f"{+(a)}")
print(f"{-(a)}")
print(f"{~(a)}")
print(f"{not(a)}")
```

```python
a = 10
print(+a)
print(-a)
```

```python
print(not a)
```

```python
a = 0
print(0, bool(a), not a)
a = 1
print(1, bool(a), not a)
a = 10 
print(10, bool(a), not a)
```
