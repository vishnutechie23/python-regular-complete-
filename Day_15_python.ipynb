```python
def add (a,b):
    return a + b 
def add2 ( a:int ,b : int ) -> int:
    return a + b 
```

```python
a = int(input("Enter the first number"))
b = int(input("Enter the second number"))

total = add(a,b)

print(f"The sum of {a } & {b} is {total}")
```

```python
x = int(input("Enter the first number"))
y = int(input("Enter the second number"))

total = add2(x,y)

print(f"The sum of {x} & {y} is {total}")
```

```python
def is_prime(n) :
    if n < 2 :
        return False
    
    for i in range(2, n ):
            if n % i == 0:
                return False 
            
    return True

```

```python
x = int(input("Enter the number"))



print(fthe number {x} is {is_prime}")

```
