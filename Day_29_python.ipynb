#### isdigit, isalpha, isalnum 

```python
s1 ,s2, s3 = "123", "abc", "abc123"
print(f"String : {s1} \nNumeric : {s1.isnumeric()}\nAlphabetic : {s1.isalpha()}\nAlphaNumeric : {s1.isalnum()}\n")
print(f"String : {s2} \nNumeric : {s2.isnumeric()}\nAlphabetic : {s2.isalpha()}\nAlphaNumeric : {s2.isalnum()}\n")
print(f"String : {s3} \nNumeric : {s3.isnumeric()}\nAlphabetic : {s3.isalpha()}\nAlphaNumeric : {s3.isalnum()}\n")
```

```python
new_string = "abc123!@#"
print(f"String : {new_string}\nNumeric : {new_string.isnumeric()}\nAlphabetic : {new_string.isalpha()}\nAlphaNumeric : {new_string.isalnum()}")
```

```python
# QS (wipro) : given string, count number of alphabets, digit and speial charcters

string = "abc123!@#   "

a,d = 0,0 
for char in string:
    if char.isalnum():
        if char.isnumeric():
            d += 1
        else:
            a += 1
    
print(f"String : {string}\nAlphabets : {a}\nDigits : {d}\nSpecial characters : {len(string) -a -d}")

```

```python
def isPalindrome(n:int) -> bool:
    n = str(n) #Typecasting
    return n == n[::-1] 
```

```python
isPalindrome(1001)
```

```python
isPalindrome(123454323)
```
