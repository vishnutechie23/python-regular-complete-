```python
# Qs 21 : hollow pyramid 
n = int(input("Enter a number to create a hollow pyramid :")) #Take input from user 
for i in range(n): # outer loop for rows 
    for sp in range(n-i-1): # to create spaces 
        print(" ",end=" ")
    for st in range(2*i+1): # for stars 
        # print(f"{i,st}",end=" ") # to check co-ordinates and set logic 
        if st == 0 or i == n-1 or st == 2*i : #This is for right,left and bottom line 
            print("*",end=" ") # print star else space 
        else:
            print(" ",end=" ")
    print()   
```

```python
# Qs 22 : Diamond pattern 
n = int(input("Enter a nummber to craete a diamond pattern : "))
for i in range(n):
    for sp in range(n-i-1): #Upper left space logic 
        print(" ",end=" ")
    for st in range(2*i+1): #upper half triangle pattern logic  
        print("*",end=" ")
    print()
for j in range(n-1): # do n-1 becuz we need one less row on lower trianlge so it will be perfect diamond pattern 
    for sp in range(j+1): # lower left space logic 
        print(" ",end=" ")
    for st in range(((n-j-1)*2)-1): # Lower inverted trianlge logic 
        print("*",end=" ")
    print()

```
