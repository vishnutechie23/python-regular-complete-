```python
# Qs 33 : Pascal's Triangle
n = 9
prev = []

for i in range(n):
    # spaces for centering
    for sp in range(n - i):
        print(" ", end=" ")

    curr = []
    for j in range(i + 1):
        if j == 0 or j == i:
            curr.append(1)
        else:
            curr.append(prev[j-1] + prev[j])

    for val in curr:
        print(val, end="   ")

    print()
    prev = curr 
 
```

```python
n = 5
mat = [[0]*n for _ in range(n)]

top, bottom = 0, n-1
left, right = 0, n-1
num = 1

while top <= bottom and left <= right:

    # left â†’ right
    for j in range(left, right+1):
        mat[top][j] = num
        num += 1
    top += 1

    # top â†’ bottom
    for i in range(top, bottom+1):
        mat[i][right] = num
        num += 1
    right -= 1

    # right â†’ left
    if top <= bottom:
        for j in range(right, left-1, -1):
            mat[bottom][j] = num
            num += 1
        bottom -= 1

    # bottom â†’ top
    if left <= right:
        for i in range(bottom, top-1, -1):
            mat[i][left] = num
            num += 1
        left += 1

for row in mat:
    print(*row)
```

```python
# Qs 41 : number spiral matrix 
n = 3
a = [[0]*n for _ in range(n)]

top = 0
bottom = n - 1
left = 0
right = n - 1
num = 1

while top <= bottom and left <= right:

    # left â†’ right
    for j in range(left, right + 1):
        a[top][j] = num
        num += 1
    top += 1

    # top â†’ bottom
    for i in range(top, bottom + 1):
        a[i][right] = num
        num += 1
    right -= 1

    # right â†’ left
    if top <= bottom:
        for j in range(right, left - 1, -1):
            a[bottom][j] = num
            num += 1
        bottom -= 1

    # bottom â†’ top
    if left <= right:
        for i in range(bottom, top - 1, -1):
            a[i][left] = num
            num += 1
        left += 1

for row in a:
    print(*row)
```

```python
# 2-D patter using recursion 
def pattern2(row, col):
    if row == 0:
        return

    def print_col(c):
        if c == 0:
            return
        print("*", end=" ")
        print_col(c - 1)

    print_col(col)
    print()
    pattern2(row - 1, col)

pattern2(3, 4)
```

```python
# Qs 49 : pattern using Bitwise logic 
# bitwise & 
n = 4
for i in range(n):
    for j in range(n):
        print((i + j) & 1, end=" ")
    print()
print()

# bitwise xor (^)
n = 4
for i in range(n):
    for j in range(n):
        print(i ^ j & 1, end=" ")
    print()
print()
# bitwise or | 
n = 4
for i in range(n):
    for j in range(n):
        print(i | j, end=" ")
    print()
print() 
# bitwise AND 
n = 4
for i in range(n):
    for j in range(n):
        if (i & j) == 0:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
print()
# bitwise shift (<<)
n = 4
for i in range(n):
    for j in range(n):
        print(1 << j, end=" ")
    print()
```
