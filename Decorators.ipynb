```python
def isPrime(n : int) -> bool:
    if n < 2:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False 
    else:
        return True 
```

```python
def getFactors(n : int) -> List[int]:
    li = [] 
    for i in range(1, n + 1):
        if n % i == 0:
            li.append(i) 
        
    return li 
```

```python
def getPrimeFactors(li : List[int]) -> List[int]:
    newli = [] 

    for item in li:
        if isPrime(item):
            newli.append(item) 
        
    return newli
```

```python
def final(n :int, f1, f2):
    li = f1(n) 
    newli = [] 
    for item in li:
        if f2(item):
            newli.append(item) 
    
    return newli 
```

```python
n = int(input()) 

li = getFactors(n) 
newli = getPrimeFactors(li) 

print(f"Prime Factors of {n} is {newli}")
```

```python
n = int(input()) 

newlist = final(n, getFactors, isPrime)

print(f"Prime Factors of {n} is {newlist}")
```

```python
class SavingAccount:
    # Class Attribute
    minimum_account_balance = 5000 
    account_type = "Savings Account"
    interest_rate = 5.0
    bank_name = "HDFC Bank" 
    ifsc_code = "HDFC" + "".join([str(item) for item in np.random.randint(0, 10, 5)])
    fd_interest_rate = 5
    fd_fine_percentage = 10 


    # Instance Attribute 
    def __init__(self, name, age, gender, email, contact, address, balance, fd_amount = 0):
        self.name = name 
        self.age = age 
        self.gender = gender 
        self.contact = contact 
        self.email = email 
        self.address = address 
        self.balance = balance 
        self.fd_amount = fd_amount
        self.account_number = "".join([str(item) for item in np.random.randint(0, 10, 15)])
        self.due = 0 
        self.fd_time_period = 0
        self.isFdMatured = False  
        self.loan_limit = 0  
    
    def deposit(self, amount):
        self.balance += amount 
        print(f"{amount} deposited in {self.account_number} successfully\nAvl Bal : {self.balance}") 
    
    def withdraw(self, amount):
        if self.balance - amount < SavingAccount.minimum_account_balance:
            print(f"Insufficient Funds") 
        else:
            self.balance -= amount
            print(f"{amount} Debited from {self.account_number} successfully\nAvl Bal : {self.balance}\n")


    def displayDetails(self):
        print(f"Bank Name : {SavingAccount.bank_name}\nIFSC Code : {SavingAccount.ifsc_code}\nAccount Holder Details\nName : {self.name}\nAccount Number : {self.account_number}\nAge : {self.age}\nGender : {self.gender}\nEmail : {self.email}\nContact : {self.contact}\nAddress : {self.address}\nBank Balance : {self.balance}")

    def applyLoan(self):
        data = int(input("Please Select From the options Below\n1. Check Loan Limit\n2. Get Loan"))

        if data == 1:
            self.loan_limit = self.balance * 200000 // 10000
            print(f"Your Loan Limit : {self.loan_limit}") 
        elif data == 2:
            amount = int(input("Enter the ammount you want to apply loan for : "))
            self.loan_limit -= amount 
            self.due += amount 
            self.balance += amount 
            print(f"{amount} Deposited In {self.account_number} Successfully") 
        else:
            print(f"Invalid Input")

    def repayLoan(self, amount):
        if amount > self.due:
            amount -= self.due 
            self.due = 0 
            self.balance += amount 
            print(f"Your Due Payment : {self.due}\n{amount} Deposited In {self.account_number} Successfully")
        else:
            self.due -= amount
            if self.due == 0:
                print(f"Your Due Payment : {self.due}")
            else:
                print(f"{amount} Repayed to {SavingAccount.bank_name}\nYour Due Payment : {self.due}")

    def createFD(self, amount, time_period):
        self.fd_time_period = time_period
        self.fd_amount = amount 
    
    def getFdAmount(self):
        if self.isFdMatured:
            final_fd_amount = self.fd_amount * (1 + SavingAccount.fd_interest_rate * self.fd_time_period * 0.01) 
            self.balance += final_fd_amount
            print(f"{final_fd_amount} Deposited in {self.account_number} Successfully") 
        else:
            final_fd_amount = self.fd_amount * (100 - SavingAccount.fd_fine_percentage) * 0.01  
            self.balance += final_fd_amount
            print(f"{final_fd_amount} Deposited in {self.account_number} Successfully")

        self.fd_amount = 0 
        self.fd_time_period = 0 
        self.isFdMatured = False 

    def matureFD(self):
        self.isFdMatured = True 
```

```python
class Amazon:
    standard_discount = 5

    def __init__(self, name, contact, email, product, price):
        self.name = name 
        self.contact = contact 
        self.email = email  
        self.product = product 
        self.price = price 


    def displayBill(self):
        print(f'Product : {self.product}\nPrice : {self.price}')
    
    @staticmethod
    def greet():
        print(f"Thank you for visiting Amazon") 

    @classmethod
    def billing(cls):
        print(f"The Standard Discount is {cls.standard_discount}")
        cls.greet() 
        
```
