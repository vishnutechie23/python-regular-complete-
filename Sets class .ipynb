```python
# Insertion : Single Element 



data = int(input(f"Enter the number you want to add in the set\nSet = {s}"))

print(f"Data to Insert in the set : {data}")
print(f"Before Updation\nSet = {s}")
s.add(data) 
print(f"After Updation\nSet = {s}")
```

```python
# Insertion : Multiple Elements (Alternative 1 : Loop Method) 

s = {23,  4362, 35, 4, 74, 6243, 45, 734, 124, 2435, 37,26}

print(f"Original Set : {s}")

n = int(input("Enter the number of elements you want to insert in the set : "))


for i in range(n):
    data = int(input("Enter the data : "))
    s.add(data) 

print(f"Updated Set : {s}")
```

```python
s = {23, 23, 2, 34, 1, 4}

newset = {6,5,3,8,4}

print(f"s = {s}, length = {len(s)}")
s.update(newset)
print(f"s = {s}, length = {len(s)}")
```

```python
# Set Comprehension 

newset = {int(item) for item in input().split()}
print(newset) 
```

```python
s = {23, 23, 2, 34, 1, 4}

newset = {int(item) for item in input().split()}

print(f"s = {s}, length = {len(s)}")
s.update(newset)
print(f"s = {s}, length = {len(s)}")
```

```python
# Deletion 

s = {8, 2, 4,  1, 5, 10, 30, 20, 34}

print(f"Set = {s}")

# Element Wise Deletion (Index wise deletion Not possible) 

s.remove(10) 

print(f"Set = {s}")

# s.remove(100) # KeyError: 100 (Note : if key not present, then display error)

# Use the discard function 

s.discard(100) 
print(f"Set = {s}")

s.discard(34) 
print(f"Set = {s}")

# Random Element Deletion 

s.pop() 
print(f"Set = {s}")

s.pop() 
print(f"Set = {s}")

s.pop() 
print(f"Set = {s}")

# Deletion all the elements in the set 
s.clear() 

print(f"Set = {s}")

# Delete the whole set 
del s 

# print(f"Set = {s}")  # NameError: name 's' is not defined
```

```python
s1 = {1, 2, 3, 4}  
s2 = {3, 4, 5, 6} 
'''
Union : Select all the elements from both the sets 
Intersection : Select common elements from the sets
Difference : 
Symmetric_Difference :

'''

print(f"Set 1 : {s1}\nSet 2 : {s2}\nUnion : {s1.union(s2)}\nInsersection :{s1.intersection(s2)}\n{s1} - {s2} : {s1.difference(s2)}\n{s2} - {s1} : {s2.difference(s1)}\nSymmetric Difference : {s1.symmetric_difference(s2)}")
```

```python
s1 = {1, 2, 3, 4} 
s2 = {3, 4, 5, 6} 
s3 = {1, 2} 

print(f"{s1} Subset of {s2} : {s1.issubset(s2)}")
print(f"{s1} Subset of {s3} : {s1.issubset(s3)}")
print(f"{s3} Subset of {s1} : {s3.issubset(s1)}")
print(f"{s1} Superset of {s3} : {s1.issuperset(s3)}")
print(f"{s1} Not Joined to  {s2} : {s1.isdisjoint(s2)}")
print(f"{s2} Not Joined to  {s3} : {s2.isdisjoint(s3)}")
```
