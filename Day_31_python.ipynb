## Frequency in dict

```python
string = "The purpose of life is to live it, to taste experience to the utmost, to reach out eagerly and without fear for newer and richer experience." 
```

```python
dictionary = {}
for char in string:
    dictionary[char] = dictionary.get(char, 0) + 1 

dictionary  
```

```python
dictionary = {}
for char in string.split():
    dictionary[char] = dictionary.get(char, 0) + 1 

dictionary 
```
