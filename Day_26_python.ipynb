## Line seperated input 

```python
subject = ['Marathi', 'Mathematics', 'Science', 'Social Science', 'Hindi'] 
marks_list = [] 
for i in range(len(subject)):
    marks = int(input(f"Enter the Marks for {subject[i]} : "))
    marks_list.append(marks) 

print(f"Percentage : {sum(marks_list)/len(marks_list) : 0.2f}%")   
```
