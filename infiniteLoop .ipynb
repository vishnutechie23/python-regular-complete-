# Infinite loop


```python
# Syntax

# while True:
#     Logic
```

```python
# Print number from 1 to 10 using an infinite loop

a = 1

while True:
    if a > 10:
        break
    print(a, end=" ")
    a += 1
```

```python
# Print all even numbers from 1 to 10 using infinite loop

a = 1

while True:
    if a < 11:
        if a % 2 == 0:
            print(a, end=" ")
        a += 1
    else:
        break
```

```python
# Print all even numbers from 1 to 10 using infinite loop (Second method)
a = 1

while True:
    if a > 10:
        break
    else:
        if a % 2 == 0 :
            print(a, end=" ")
        a += 1 
```

```python
#  Better Version (Low Complexity)

# Print all even numbers from 1 to 10 using infinite loop

a = 1

while True:
    if a > 10:
        break
    elif a % 2 == 0:
        print(a, end=" ")
        a += 1
    else:
        a += 1
        continue
```
