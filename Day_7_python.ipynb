```python
# print even numbers from 1 to 10
start = 1
stop = 10

while start <= stop:
    if start % 2 == 0:
        print(start,end=" ")
        start +=1
    else:
        start +=1 
```

```python
# print odd numbers from 1 to 10
start = 1
stop = 10

while start <= stop:
    if start % 2 != 0:
        print(start,end=" ")
        start +=1
    else:
        start +=1 
```

```python
# print in revrese format
start = int(input("Enter a number"))
stop = 0 
a = start
while a >= stop:
    print(a)
    a -= 1 
```

```python
start = 0
stop = 100

count = start
while count <= stop:
    if count > 1:  # 0 and 1 are not prime
        is_prime = True
        i = 2
        while i <= count // 2:  # check divisibility
            if count % i == 0:
                is_prime = False
                break
            i += 1
        if is_prime:
            print(count, end="  ")
    count += 1

```
