```python
# Split Wise 

from typing import * 
```

```python
class SplitWise:
    def __init__(self, name : List[str], expense : List[int]):
        self.name = name 
        self.expense = expense 
        self.receivers = None 
        self.givers = None 
        self.neutral = None 
    
    def per_person_contribution(self):
        return sum(self.expense) / len(self.expense) 
    
    def displayDetails(self):
        for index, item in enumerate(self.name):
            print(f"{item} --> {self.expense[index]}")

    def settlement(self):
        print("-" * 80)
        print(f"{'Name':<15}{'Amount Paid':<15}{'Amount Due/Balance':<   }{'Status'}")
        print("-" * 80)

        for name, paid, balance in self.receivers:
            print(f"{name:<15}{paid:<15.2f}{balance:<20.2f}{'Will Receive'}")

        for name, paid, balance in self.givers:
            print(f"{name:<15}{paid:<15.2f}{balance:<20.2f}{'Will Give'}")

        for name, paid, balance in self.neutral:
            print(f"{name:<15}{paid:<15.2f}{balance:<20.2f}{'Neutral'}")

        print("-" * 80)
    
    def final_distribution(self):
        pass 


    def split_person(self):
        if self.receivers is None:
            self.receivers = [] 

        if self.givers is None:
            self.givers = [] 
        
        if self.neutral is None:
            self.neutral = [] 
        
        for index, item in enumerate(self.name):
            if self.expense[index] > self.per_person_contribution():
                self.receivers.append((item, self.expense[index], round(self.expense[index] - self.per_person_contribution(), 2) ))
            elif self.expense[index] < self.per_person_contribution():
                self.givers.append((item, self.expense[index], round(self.per_person_contribution() - self.expense[index], 2)))
            else:
                self.neutral.append((item, self.expense[index], 0))

        self.receivers.sort(key = lambda x : x[2]) 
        self.givers.sort(key = lambda x : x[2]) 
        self.neutral.sort(key = lambda x : x[2]) 

        print(self.receivers) 
        print(self.givers) 
        print(self.neutral)  
```

```python
final_list = ['Jack Sparrow', '25', 'Kakashi Hatake', '75', 'Harry Potter', '40', 'Iron Man' , '60', 'Shah Rukh Khan', '23'] 

name_list = [final_list[item] for item in range(0, len(final_list), 2)]
expense_list = [int(final_list[item]) for item in range(1, len(final_list), 2)]

sp = SplitWise(name_list, expense_list) 

sp.displayDetails()
```
