```python
# Class iStore 


class iStore:
    # Class Attribute
    company_name = "Apple" 
    address = "Infinity Mall Malad"
    company_categories = "Electronics" 
     
    product_price = {
        "Iphone 17 Pro Max" : 150000, 
        "Iphone 17 Pro" : 120000, 
        "Iphone 17" : 80000, 
        "IWatch 7 Series" : 30000, 
        "MacBook 10 Series" : 85000 
    }

    # Instance Attribute --> Constructor 
    def __init__(self, name, email, product_name, payment_method):
        self.name = name 
        self.email = email 
        self.product_name = product_name 
        self.payment_method = payment_method 
        self.discounted_price = self.price = iStore.product_price[self.product_name]
        self.discount = self.calcualteDiscount()
        self.isDiscountApplied = False  
    
    def calcualteDiscount(self):
        if self.price >= 120000:
            return 0.20 
        elif self.price >= 80000:
            return 0.10
        else:
            return 0.05 
        
    def displayDetails(self):
        if self.isDiscountApplied:
            print(f"Name : {self.name}\nEmail : {self.email}\nProduct Purchased : {self.product_name}\nPrice : {self.price}\nPayment Method : {self.payment_method}\nProduct Price : {self.price}\nDiscount : {self.discount * 100 : 0.0f}%\nDiscounted Price : {self.discounted_price}")
        else:
            print(f"Name : {self.name}\nEmail : {self.email}\nProduct Purchased : {self.product_name}\nPrice : {self.price}\nPayment Method : {self.payment_method}\nProduct Price : {self.price}\nDiscount : {self.discount * 100 : 0.0f}%")

        
    def applyDiscount(self):
        self.discounted_price = self.price * (1 - self.discount) 
        self.isDiscountApplied = True 
```
