## Function 
#### Single line function : lambda function 

```python
def add(a : int, b : int) -> int :
    return a + b 
```

```python
add_function = lambda a,b : a+b 
```

```python
a = int(input())
b = int(input())
print(f"a = {a}, b = {b} , a + b = {a+b}")
print(f"a = {a}, b = {b} , a + b = {add(a,b)}")
print(f"a = {a}, b = {b} , a + b = {add_function(a,b)}")
```

```python
from typing import List 
```

```python
#Function declaration and defination 
def first_index (li : List[int], n : int) -> int:
    for index in range(len(li)):
        if index == n:
            return index
    else:
        return -1 
#Taking input from user 
n = int(input("Enter a number"))
li = [int(item) for item in input("Enter the list elements : ").split()]

#function call 
index = first_index(li,n)

#Outpujt display vala code
if index == -1 :
    print(f"{n} not found in {li}")
else:
    print(f"{n} found in {li} at index {index}") 
```
