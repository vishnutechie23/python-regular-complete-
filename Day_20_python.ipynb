## Indexing 
#### 2 & 3 prefered

```python
# For First Element 
print(f"1.First element- {list[-7]}") 
print(f"2.First element- {list[0]}") 
print(f"3.First element- {list[-len(list)]}") 
```

```python
# For last element
print(f"1.Last Element - {list[6]}") 
print(f"2.Last Element - {list[-1]}") 
print(f"3.Last Element - {list[len(list)-1]}") 
```
