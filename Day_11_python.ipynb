```python
# print all even numbers between 1 and 20 

for i in range(1, 21, 1):
    # if i % 2 == 0: print(i, end = " ")  IF EVEN THEN PRINT 
    # IF ODD THEN SKIP
    if i % 2 != 0:
        continue
    
    print(i, end = " ") 
```

```python
i = 1 
while i < 21:
    if i % 2 != 0:
        i += 1 
        continue
    print(i, end = " ")  
    i += 1 
```

```python
# prime number 
n = int(input("Enter a number"))
if n < 2:
    print(f"{n} is not prime number")
else:
    for num in range(2,n):
        if n % num == 0: 
            print(f"{n} is not prime")
            break
    else:
        print(f"{n} is prime number")
```

## Nested looping

```python
for outer_loop in range(1,6):
    print(f"{outer_loop}" , end=" -->")
    for inner_loop in range (10,13):
        print(f"{inner_loop}",end=" ")
    print()
```

```python
for outer_loop in range(1,6):
    print(f"{outer_loop}" , end=" -->")
    for inner_loop in range (10,13):
        print(f"{inner_loop}",end=",")
    print()
```

```python
for i  in range(5):
    for j in range(5):
        print(f"({i + 1}  {j + 1})" , end=" ")
    print()        
```

```python
for i  in range(5):
    for j in range(5):
        print(f"( Row = {i + 1}, col={j + 1})" , end=" ")
    print()        
```
