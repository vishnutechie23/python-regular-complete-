```python
# Insertion of multiple element (Alternative 1 : loop method)
s = {1,2,3,4,5}
print(f"Original Set : {s}") 

n = int(input("Enter the number of element you want to insert into set :")) 

element = []
for i in range(n):
    data = int(input("Enter the data"))
    element.append(data) 
    s.add(data)
print(F"Number of element you entered to insert in the set : {n}")
print(f"elements you want to insert : {element}")
print(f"Updated set : {s}")
```

```python
# insertion of new set 
s = {1,2,3,4,5}

newset = {6,7,8,9}
print(f"s = {s},length = {len(s)}")
s.update(newset)
print(f"New added elements : {newset}")
print(f"s = {s}, length = {len(s)}") 
```

```python
# Set comprehensions 

newset = {int(item) for item in input().split()} 
print(newset)
```

```python
s = {12,23,45,34,56}
newset = {int(item) for item in input().split()} 

print(f"s = {s}, length = {len(s)}") 
s.update(newset)
print(f"s = {s},lenght = {len(s)}")  
```

```python
# Deletion 
s = {1,2,3,4,5,6,7,8}
print(f"set : {s}") 

# Element wise deletion (Index wise deletion not possible)
s.remove(5)
print(f"set : {s}") 

# s.remove(101) KeyError: 101 (if key not present, then display error)

# Use the discard function 

s.discard(100)
print(f"Set : {s}") 
s.discard(4)
print(f"Set : {s}") 

# Random element Deletion 
s.pop()
print(f"Set = {s}")

s.pop()
print(f"Set = {s}")

s.pop()
print(f"Set = {s}")

# Deletion of all th elements in the set 
s.clear()

print(f"Set = {s}")

# Deletion of the whole set 
del s 
# print(s)  NameError: name 's' is not defined so set delete ho gaya and variable is free now
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

'''
Union : Select all the element from the both the sets 
Intersetion : select common element from the sets 
Difference :
Symmnetric_difference :

'''

print(f"Set 1 : {s1}\nSet 2 : {s2}\nUnion : {s1.union(s2)}\nIntersection : {s1.intersection(s2)}\n{s1} - {s2} : {s1.difference(s2)}\n{s2} - {s1} : {s2.difference(s1)}\nSymmetric Differnce : {s1.symmetric_difference(s2)}")
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7}
s3 = {1,2} 

print(f"{s1} Subset of {s2} : {s1.issubset(s2)}")
print(f"{s3} Subset of {s1} : {s3.issubset(s1)}")
print(f"{s2} Subset of {s1} : {s2.issubset(s1)}")
print(f"{s1} superset of {s2} : {s1.issuperset(s2)}")
print(f"{s1} superset of {s3} : {s1.issuperset(s3)}")
print(f"{s1} Not joined to {s2} : {s1.isdisjoint(s2)}")
print(f"{s2} not joined to {s3} : {s2.isdisjoint(s3)}")
print(f"{s1} Subset of {s2} : {s1.intersection(s2)}") 
```
