## DELETION in list 

#### Index -> pop , agar sirf pop() likha means : last index deletion else pop(2) : delete at particular index 
#### Element -> remove 

```python
# Index Based 
li = [8, 2, 4, 1, 5, 20, 7] 

print(f"Original List : {li}")

# Removing the last Element : 

removed_number = li.pop() 

print(f"Updated List 1: {li}")

# Removing on the basis of index (0th index) 

removed_data = li.pop(3) 

print(f"Updated List 2: {li}")

```

```python
# Deletion - Element 

li = [8, 2, 4, 1, 5, 2, 9] 
print(li) 

li.remove(4) 
print(li) 

# li.remove(1000)  : shows error bcuz element hai hi nhi list mai 
li.remove(5) 
print(li) 
```
