## if - else clause

```python

# Check whether a Person can vote or not 
# take input from user 
age = int(input("Enter the age :"))

# decision making 
if age >= 18:
    print(f"you can vote Since you are {age} years old")
else:
    print(f"you can't vote Since you are {age} years old")
    
```

```python
# check whether a number is even or odd 

number = int(input("Enter the number :"))

if number % 2 == 0:
    print(f"{number} is Even")
else:
    print(f"{number} is Odd")
```

```python
# check whether  a number is positive, negative or zero

n = int(input())

if n < 0:
    print(f"{n} is Negative")
elif n > 0:
    print(f"{n} is Positive")
else:
    print(f"{n} is Zero")
```

```python
# check wheather a number is even or odd ( both positive or negative)

n = int(input())

if n > 0:
    if n % 2 == 0:
        print(f"{n} = Positive Even")
    else:
        print(f"{n} = positive Odd")

elif n < 0:
    if n % 2 == 0:
        print(f"{n} = Negetive Even")
    else:
        print(f"{n} = Negetive Odd")
        
else:
    print(f"{n} = Zero")
```

```python
## Check whether  a number is positive, negative or zero and check it is divisible by 5 or not 

a = int(input())

if a > 0 :
    if a % 5 == 0:
        print(f"{a} is positive and divisible by 5")
    else:
        print(f"{a} is Positive and not divisible by 5")
elif a < 0:
    if a % 5 == 0:
        print(f"{a} is Negetive and divisible by 5") 
    else:
        print(f"{a} is Negetive and not divisible by 5")
else:
    print(f"{a}is Zero") 
```

## Grading system


```python
M = int(input("Enter your Marks :"))
if M >= 90:
    print(f"Your marks are {M} and You have A grade")
elif 90 > M > 80 :
    print(f"Your marks are {M} and You have B grade")
elif 80 > M > 70:
    print(f"Your marks are {M} and You have c grade")
elif 70 > M > 50:
    print(f"Your marks are {M} and You have D grade")
elif 35 > M > 50:
    print(f"Your marks are {M} and You have E grade")    
else:
    print(f"Your marks are {M} and You are Fail")
```

## Grading system with percentage

```python


M = int(input("Enter your Marks :"))

per = M / 600 * 100 
if per >= 90:
    print(f"Your marks are {per} % and You have A grade")
elif 90 > per > 80 :
    print(f"Your marks are {per} % and You have B grade")
elif 80 > per > 70:
    print(f"Your marks are {per} % and You have c grade")
elif 70 > per > 50:
    print(f"Your marks are {per} % and You have D grade")
elif 35 > per > 50:
    print(f"Your marks are {per} % and You have E grade")    
else:
    print(f"Your marks are {per} % and You are Fail")
    
```

```python
## calculator 
a = int(input("Enter a first number"))
activity = (input("enter what kind of caculations you want to perform ?"))
b = int(input("Enter a second number"))

if  activity == "+":
    print(f"{a} + {b} is {a+b}")
elif activity == "-":
    print(f"{a} - {b} is {a-b}")
elif activity == "*":
    print(f"{a} * {b} is {a*b}")
elif activity == "/":
    print(f"{a} / {b} is {a/b}")
else :
    print(f"{activity} please use only +,-,*,/ only")
```
