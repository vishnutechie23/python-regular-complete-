# Encapsulation 

### It is the process of binding attributes(variable) and methods(functions) in a single unit called class and restricting the access of data using access modifiers. 

```python
class GovernmentOffice:
    # Attributes 
    
    # Class Attribute 
    office_address = "Borivali MahanagerPalika" 
    office_hours = ["9:00:00", "17:00:00"] 
    lunch_time = ["14:00:00", "16:00:00"] 

    # Instance Attribute 
    def __init__(self, name, age, addressProof, addharCard, panCard, rationCard):
        self.name = name 
        self.age = age 
        self.addressProof = addressProof
        self.addharCard = addharCard
        self.panCard = panCard 
        self.rationCard = rationCard         
        self.isRegistryDone = False 
        self.registry_status = "Not Registred"

    # Methods 
    # Instance Methods 
    def getPropertyProof(self):
        print(f"Property Registration") 
        print(f"Name : {self.name}\nAddhar Number : {self.addharCard}\nPan Number : {self.panCard}\nAddress : {self.addressProof}\nRegistry Status : {self.registry_status}")

    def FinalRegistration(self):
        self.isRegistryDone = True 
        self.registry_status = "Registred" 
```

# Protected 

```python
import numpy as np 
import random 
```

```python
class Government_Office:
    # Attributes 
    # Protected Instance Attributes 
    def __init__(self, name, age, gender, contact, address):
        self._name = name 
        self._age =age 
        self._gender = gender 
        self._contact = contact 
        self._address = address 
        self._addhar_number = None 
        self._pan_number = None 
        self._voter_id = None 

    # Methos
    # Instance Methods

    # Protected Instance Method 
    def _getAddharCard(self):
        print(f"Addhar Card\nName : {self._name}\nAge : {self._age}\nGender : {self._gender}\nContact Number : {self._contact}\nAddress : {self._address}\nAddhar Number : {self._addhar_number}") 
    
    # Protected Instance Method 
    def _getPanCard(self):
        print(f"Pan Card\nName : {self._name}\nAge : {self._age}\nGender : {self._gender}\nContact Number : {self._contact}\nAddress : {self._address}\nPan Number : {self._pan_number}")

    # Protected Instance Method 
    def _getVoterId(self):
        print(f"Voter Id Card\nName : {self._name}\nAge : {self._age}\nGender : {self._gender}\nContact Number : {self._contact}\nAddress : {self._address}\nVoter Id Number : {self._voter_id}")

    # Protected Instance Method 
    def _generateAddharNumber(self):
        self._addhar_number = "".join([str(item) for item in np.random.randint(0, 10, 4)]) + " " + "".join([str(item) for item in np.random.randint(0, 10, 4)]) + " " + "".join([str(item) for item in np.random.randint(0, 10, 4)])

    # Protected Instance Method 
    def _generatePanNumber(self):
        self._pan_number = "".join(chr(item) for item in np.random.randint(ord("A"), ord("Z"), 5)) + "".join(str(item) for item in np.random.randint(0, 10, 4)) + chr(np.random.randint(ord("A"), ord("Z")))
```

```python
# Agent (Child Class) , Government_Office (Parent Class) 
class Agent(Government_Office):
    def __init__(self, name, age, gender, contact, address):
        # super().__init__(name, age, gender, contact, address)
        Government_Office.__init__(self, name, age, gender, contact, address)   

    # Public Instance Method
    def getAddharCard(self):
        self._getAddharCard() 
    
    # Public Instance Method
    def getPanCard(self):
        self._getPanCard() 

    # Public Instance Method    
    def getVoterID(self):
        self._getVoterId()
    
    # Public Instance Method
    def getPassport(self):
        pass 
```

# Private  

```python
class BankAccount:
    def __init__(self, name, age,  contact, email, balance):
        self.__name = name 
        self.__age = age 
        self.__contact = contact 
        self.__email = email 
        self.__balance = balance 

    # Private Instance Method 
    def __deposit(self, ammount):
        self.__balance += ammount 
    
    def __withdraw(self, ammount):
        if self.__balance <= ammount:
            return "Insufficient Bank Balance" 
        else:
            self.__balance -= ammount
    
    def __displayDetails(self):
        print(f"Name : {self.__name}\nAge : {self.__age}\nContact : {self.__contact}\nEmail : {self.__email}\nBank Balance : {self.__balance}")

    def deposit(self, ammount):
        self.__deposit(ammount) 

    def withdraw(self, ammount):
        self.__withdraw(ammount)

    # Public Instance Method 
    def displayDetails(self):
        self.__displayDetails() 
```
