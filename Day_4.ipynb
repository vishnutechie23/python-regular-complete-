### Operators 

#### Identity Operator 

```python
a = 10 
b = 20 

print(f"{a} == {b} : {a == b}")
print(f"{a} IS {b} : {a is b}")
print(f"Memory Location of {a} : {id(a)}\nMemory Location of {b} : {id(b)}")
```

#### Membership Operator  

```python
li = [8, 2, 4, 1, 5]

print(f"List : {li}") 
print(f"1 in {li} : {1 in li}")
print(f"10 in {li} : {10 in li}")
print(f"100 not in {li} : {100 not in li}")
print(f"8 not in {li} : {8 not in li}")
```

#### Unary Operator 

```python
a = 10 

print(+a) 
print(-a) 
```

```python
print(~a) 
```

```python
print(not a) 
```

```python
a = 0 

print(a, bool(a), not a)  
```
