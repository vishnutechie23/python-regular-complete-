```python
n = int(input("Enter a number"))
for i in range(n):
  for sp in range(n - i -1):
    print(" ",end=" ")
  for st in range(2*i+1):
      if i == n-1 or st == 0:
        print("*",end=" ")
      elif st == i*2 :
        print("*",end=" ")
      else:
        print(" ",end=" ")
  print()
```

```python
# Sum of n number 
n = int(input("Enter a number"))
st = 1
total = 0
while st < n + 1:
    total = total + st
    st += 1
print(total)
```

```python
def sum (n :int) -> int:
  if n == 0:
    return 0
  small_ans = sum(n-1)
  final_ans = small_ans + n
  return final_ans

sum(20)
```
